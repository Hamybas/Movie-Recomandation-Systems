# Content-Based Filtering

```
import pandas as pd
```

```
movies = pd.read_csv("movies.csv", sep=',')
```

```
movies
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
tfindf = TfidfVectorizer(stop_words='english')
```

```
movies['overview'] = movies['overview'].fillna('')
```

```
tfidf_matrix = tfindf.fit_transform(movies['overview'])
```

```
pd.DataFrame(tfidf_matrix.toarray(), columns=tfindf.get_feature_names_out())
```

```
tfidf_matrix.shape
```

### Similatiry matrix

```
from sklearn.metrics.pairwise import linear_kernel

similarity_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```
pd.DataFrame(similarity_matrix)
```

### Find the most similar movies to a certain movie

```
movie_title = "John Carter"
```

```
idx = movies.loc[movies['title'] == movie_title].index[0]
idx
```

```
scores = list(enumerate(similarity_matrix[idx]))
scores
```

```
sorty = sorted(scores, key=lambda x: x[1], reverse=True)
sorty
```

```
movies_indices = [tpl[0] for tpl in sorty[1:4]]
movies_indices
```

```
movies["title"].iloc[movies_indices]
```

```
def similar_movies(movie_title, nr_movies):
    idx = movies.loc[movies['title'] == movie_title].index[0]
    scores = list(enumerate(similarity_matrix[idx]))
    sorty = sorted(scores, key=lambda x: x[1], reverse=True)
    movies_indices = [tpl[0] for tpl in sorty[1:nr_movies+1]]
    sim_titles = movies["title"].iloc[movies_indices]
    return sim_titles
```

```
similar_movies("Kung Fu Panda 3", 5)
```
